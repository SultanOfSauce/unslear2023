## Exercise 1: Gaussian

Generate a dataset drawn from two independent Gaussians, each of which represents a class. You are free to select the number of desired datapoints, the mean and the variance of the two distributions.
Flip randomly a percentage of the labels to introduce some noise and plot the results.


```python
def double_class_gaussian(n_points, mean1, mean2, var1, var2, amt_noise):
  total_points = 2*n_points
  classes = np.zeros(total_points)
  pts1 = np.random.multivariate_normal(mean=mean1, cov = var1, size = n_points)
  pts2 = np.random.multivariate_normal(mean=mean2, cov = var2, size = n_points)

  classes[n_points:] = 1

  noise = np.full(total_points, False)
  noise[0:total_points*amt_noise//100] = True
  noise = np.random.choice(noise, size = total_points, replace=False)

  classes[noise] = np.abs(1-classes[noise])

  data = np.append(pts1,pts2,axis=0)

  return data,classes

v1 =[[0,5],
 [5,0]]

v2 =[[0,.1],
 [.1,0]]

m1 = [1,1]
m2 = [10,10]


data, classes = double_class_gaussian(100, m1, m2, v1, v2, 10)

print(data[:5])
print(classes[:5])

gauss = plt.subplot()
gauss.axis("equal")
gauss.scatter(data[:,0], data[:,1], c=classes)



```

## Exercise 2:
Swiss Roll

```python
def create_swiss_roll(n_points):

  phi = np.random.uniform(low=1.5 * np.pi, high=4.5 * np.pi, size=n_points)
  psi = np.random.uniform(low=0, high=10, size=n_points)

  x = phi * (np.cos(phi))
  y = phi * (np.sin(phi))
  z = psi

  return x,y,z


x,y,z = create_swiss_roll(1000)

fig = plt.figure(figsize=(16,8))
roll = fig.add_subplot(1,2,1,projection='3d')
roll.scatter(x,y,z)
roll_side = fig.add_subplot(1,2,2)
roll_side.scatter(x,y)
```

## Exercise 3
Klein bottle

```python
def create_klein_bottle(n_points):

  u = np.random.uniform(low=0, high=np.pi, size=n_points)
  v = np.random.uniform(low=0, high=2*np.pi, size=n_points)

  x = (-2/15) * cos(u) * (3 * cos(v) - 30 * sin(u) + 90 * np.power(cos(u), 4) * sin(u) - 60 * np.power(cos(u), 6) * sin(u) + 5 * cos(u) * cos(v) * sin(u))
  y = (-1/15) * sin(u) * (3 * cos(v) - 3 * np.power(cos(u), 2) * cos(v) - 48 * np.power(cos(u),4) * cos(v) + 48 * np.power(cos(u), 6) * cos(v) - 60 * sin(u) + 5 * cos(u) * cos(v) * sin(u) - 5 * np.power(cos(u), 3) * cos(v) * sin(u) - 80 * np.power(cos(u),5) * cos(v) * sin(u) + 80 * np.power(cos(u),7) * cos(v) * sin(u))
  z = (2/15) * (3 + 5 * cos(u) * sin(u)) * sin(v)
  return x,y,z


x,y,z = create_klein_bottle(5000)

fig = plt.figure(figsize=(16,8))
roll = fig.add_subplot(1,2,1,projection='3d')
roll.scatter(x,y,z, c=z)
roll_side = fig.add_subplot(1,2,2)
roll_side.scatter(x,y,c=z)
```
